```python
cp = pd.get_dummies(data['cp'],prefix='cp',drop_first=True)
```

```python
restecg = pd.get_dummies(data['restecg'],prefix='restecg',drop_first=True)
```

```python
slope = pd.get_dummies(data['slope'], prefix='slope',drop_first=True)
```

```python
ca = pd.get_dummies(data['ca'],prefix='ca',drop_first=True)
```

```python
thal = pd.get_dummies(data['thal'],prefix='thal',drop_first=True)
```

```python
df = pd.concat([data,cp,restecg,slope,ca,thal],axis=1)
```

```python
df.drop(['cp','restecg','slope','ca','thal'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
# Spliting
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('target',axis=1)
y = df.target
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
# modeling
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100, criterion='entropy')
```

```python
#Traning
rfc.fit(x_train,y_train)
```

```python
# prediction
y_pred = rfc.predict(x_test)
```

```python
# model evolution
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_pred))
print('\n')
print(confusion_matrix(y_test, y_pred))
print('\n')
print(round(accuracy_score(y_test, y_pred),2))
```
